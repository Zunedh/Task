```python
third_quartile=hr['MonthlyIncome'].quantile(0.75)
first_quartile=hr['MonthlyIncome'].quantile(0.25)
```

```python
max_val=third_quartile+1.5*(third_quartile-first_quartile)
min_val=first_quartile-1.5*(third_quartile-first_quartile)
```

```python
res=hr[hr['MonthlyIncome']>max_val]
res.head()
```
